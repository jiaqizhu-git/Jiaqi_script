```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cs_df_all = pd.read_csv('/home/daisy/Documents/Research/Rovient/roivant_data/chemical_shifts.all.csv', header = None)
cs_df_all = cs_df_all.drop(labels= range(2716, 2721), axis = 0)

cs_df_all
```

```python
cluster_assignment = np.loadtxt('/home/daisy/Documents/Research/Rovient/roivant_data/frame_cluster_assign_final.txt')
frame = cluster_assignment[:,0]
cluster = cluster_assignment[:,2].astype(int)
clusters = np.column_stack((frame,cluster))

frames_cl = []
for i in range(4):
    frame = np.where(cluster==i)[0]
    frames = np.concatenate(([-2,-1], frame))
    frames_cl.append(frames)
frames_cl
```

```python
def get_cluster_df(df, cluster, frames):
    cl = df[frames[cluster]+2]
    cl['average'] = cl.iloc[:,2:].mean(axis = 1)

    cs_shift = pd.DataFrame({'residue':cl[0], 'atom':cl[1], 'shift':cl['average']})
    CA_df_ave = cs_shift[(cs_shift == 'CA').any(axis=1)]
    CB_df_ave = cs_shift[(cs_shift == 'CB').any(axis=1)]
    HA_df_ave = cs_shift[(cs_shift == 'HA').any(axis=1)]
    C_df_ave = cs_shift[(cs_shift == 'C').any(axis=1)]
    H_df_ave = cs_shift[(cs_shift == 'H').any(axis=1)]
    N_df_ave = cs_shift[(cs_shift == 'N').any(axis=1)]
    
    CA_df_ave.set_index('residue', inplace=True, drop = False)
    CB_df_ave.set_index('residue', inplace=True, drop = False)
    HA_df_ave.set_index('residue', inplace=True, drop = False)
    C_df_ave.set_index('residue', inplace=True, drop = False)
    H_df_ave.set_index('residue', inplace=True, drop = False)
    N_df_ave.set_index('residue', inplace=True, drop = False)

    cs_all_ave = pd.DataFrame({'residue':CA_df_ave['residue'], 'CA':CA_df_ave['shift'], 'CB':CB_df_ave['shift'], 'HA':HA_df_ave['shift'], 'C':C_df_ave['shift'], 'H':H_df_ave['shift'], 'N':N_df_ave['shift']})
    cs_all_ave = cs_all_ave[0:-1]
    return cs_all_ave
```

```python
cs_cl1_ave = get_cluster_df(cs_df_all, 0, frames_cl)
cs_cl2_ave = get_cluster_df(cs_df_all, 1, frames_cl)
cs_cl3_ave = get_cluster_df(cs_df_all, 2, frames_cl)
cs_cl4_ave = get_cluster_df(cs_df_all, 3, frames_cl)
cs_cl4_ave
```

```python
def generate_weight(n):
    weights = []
    weight = []
    interval = 100*n
    num_range = np.linspace(0,100,int(100/interval+1))
    for w1 in num_range:
        for w2 in num_range:
            for w3 in num_range:
                w4 = 100-w1-w2-w3
                if w4 < 0: 
                    continue
                weight.append(w1/100)
                weight.append(w2/100)
                weight.append(w3/100)
                weight.append(w4/100)
                weights.append(weight)
                weight = []
    return weights
    
all_weights = generate_weight(0.05)
```

```python
outdir = '/home/daisy/Documents/Research/Rovient/cluster_reweight_data/'
for weight in all_weights:
    name = str(int(weight[0]*100)) +'_'+ str(int(weight[1]*100)) +'_'+ str(int(weight[2]*100)) +'_'+ str(int(weight[3]*100)) 
    cs_cl1_reweighted = cs_cl1_ave[['CA', 'CB', 'HA', 'C', 'H', 'N']]*weight[0]
    cs_cl2_reweighted = cs_cl2_ave[['CA', 'CB', 'HA', 'C', 'H', 'N']]*weight[1]
    cs_cl3_reweighted = cs_cl3_ave[['CA', 'CB', 'HA', 'C', 'H', 'N']]*weight[2]
    cs_cl4_reweighted = cs_cl4_ave[['CA', 'CB', 'HA', 'C', 'H', 'N']]*weight[3]
    cs_cl1_reweighted.to_csv(outdir + 'c1_' + name + '.csv')
    cs_cl2_reweighted.to_csv(outdir + 'c2_' + name + '.csv')
    cs_cl3_reweighted.to_csv(outdir + 'c3_' + name + '.csv')
    cs_cl4_reweighted.to_csv(outdir + 'c4_' + name + '.csv')
```
